# K - Nearest Neighbours classifier

### Its a non-parametric classifier

###### Parametric classifiers such as Linear Regression, first fit a set of optimal parameters from a set of know observations, and then use those parameters to predict new observations
##### Non parametric classifiers just look at the known observations and compute the predictions from there (theres nothing to fit before hand)

# Dataset

```python
from sklearn.datasets import load_iris

x, y = load_iris(return_X_y=True, as_frame=True)

plt.scatter(x.iloc[:, 0][y == 0], x.iloc[:, 1][y == 0])
plt.scatter(x.iloc[:, 0][y == 0], x.iloc[:, 1][y == 1])
plt.scatter(x.iloc[:, 0][y == 0], x.iloc[:, 1][y == 2])

plt.legend(['setosa','virginica','versicolor'])
plt.show()
```

# Dummie Variables

```python
y = pd.get_dummies(y)
y.columns = ['setosa', 'virginica', 'versicolor']
y
```

# Κ- Nearest Method + Classification (Iris-DataSet)

```python
import numpy as np
import pandas as pd
import seaborn as sns
import warnings
from sklearn import preprocessing
from matplotlib import pyplot as plt
from sklearn.datasets import load_iris
warnings.filterwarnings('ignore')

    
```

```python
iris_load = load_iris()
iris = pd.DataFrame(iris_load.data, columns = iris_load.feature_names)
iris.head()
```

```python
[col for col in iris.columns] # Lets just look at the columns title
```

```python
iris['class'] = pd.Series(iris_load.target)
iris.head()
```

```python
[col for col in iris.columns]
```

```python
features = ['sepal length (cm)','sepal width (cm)','petal length (cm)','petal width (cm)']
```

```python
flower_class = ['class']
```

```python
x = iris[features]
```

```python
# x=preprocessing.scale(x)
```

```python
y =iris [flower_class]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y, random_state=1)
```

```python
from sklearn.neighbors import KNeighborsClassifier
KNN = KNeighborsClassifier(n_neighbors =5)
KNN.fit(x_train, y_train)
```

```python
y_pred = KNN.predict(x_test)
y_predall = KNN.predict(x) # this is just for plotting the entire data comparison later
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test,y_pred))
```

### Accuracy check

```python
from sklearn.metrics import accuracy_score

acc = []

for i in range(1,10):
    

    KNN = KNeighborsClassifier(n_neighbors=i)
    KNN.fit(x_train,y_train)
    Kpredict = KNN.predict(x_test)
    accuracy = accuracy_score(y_test, Kpredict)
    acc.append(accuracy)

print(acc)
```

```python
plt.figure(figsize=(10,8))
plt.plot(acc, marker='o', markersize = 10)
plt.title('Accuracy vs K number')
plt.xlabel('K number')
plt.ylabel('Accuracy')
```
